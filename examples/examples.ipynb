# Načtení knihovny

```python
import mandelbrot as mb
import numpy as np
# testovací vykreslování 
import matplotlib.pyplot as plt
```

# Generace množin

Vypočtení počtu operací před výskokem pro Mandelbrotovu množinu.

Řádek určuje reálnou část, zatímco sloupec určuje imaginární složku.
Pro klasické vykreslení množiny je stačí množinu transponovat.

```python
#vytvoř výstupní množinu pro mandelbrota 
man = np.zeros((1024, 1024), dtype=np.int32)
#vygeneruj mandelbrota 
mb.mandelbrot(-0.5, 3 + 3j, 100, man)
# vykresli pravý obsah matice
plt.imshow(man)
plt.show()
# vykresli transponovaný obsah matice
plt.imshow(man.T)
plt.show()
```

Vypočti počty operací před výskokem pro Juliovu množinu.

```python
#vytvoř výstupní množinu pro juliovu množinu 
jul = np.zeros((1024, 1024), dtype=np.int32)
#vygeneruj Juliovu množinu 
mb.julia_set(0.0, 3 + 3j, -0.8 + 0.156j, 100, jul)
# vykresli pravý obsah matice
plt.imshow(jul)
plt.show()
# vykresli transponovaný obsah matice
plt.imshow(jul.T)
plt.show()
```

# Zbarvení množin

Vizualizace množiny podle barevné mapy z knihovny `matplotlib`

```python
#vizualizuj ji 
mb.visual(man, 100, -0.5, 3 + 3j, "plasma")
#vícero barevných map
mb.visual(man, 100, -0.5, 3 + 3j, "viridis")
mb.visual(man, 100, -0.5, 3 + 3j, "jet")
mb.visual(man, 100, -0.5, 3 + 3j, "cividis")
```

To samé pro Juliovu množinu.

```python
#vizualizuj juliovu množinu 
mb.visual(jul, 100, -0.5, 3 + 3j, "plasma")
#vícero barevných map
mb.visual(jul, 100, -0.5, 3 + 3j, "viridis")
mb.visual(jul, 100, -0.5, 3 + 3j, "jet")
mb.visual(jul, 100, -0.5, 3 + 3j, "cividis")
```

# Vykreslovací aplikace

V knihovně je i vizualizační aplikace pro Mandelbrotovu a Juliovu množinu.

Aplikace se spouští tímto příkazem:

```python
mb.init_app()
```

Ovládání aplikace:

- Levé tlačítko myši a tah - pohyb po obrazovce 
- Kolečko myši - přiblížení a oddalování 
- A, D - změna barevné mapy
- T - přepínání automatického obnovení
- R - obnoví množinu
- G, H - změna počtu iterací
- O - resetování pohledu

- J - přepínání mezi Mandelbrotovou a Juliovou množinou
- Shift + Levé tlačítko myši a tah - změna hodnoty C u Juliovy množiny
