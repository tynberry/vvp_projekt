# Generace mandelbrota

Načtení potřebných knihoven:

```python
# načtení knihovny
import mandelbrot as mb
# testovací vykreslování 
import matplotlib.pyplot as plt
```

Vypočti a vykresli Mandelbrotovu množinu:

```python
#vygeneruj mandelbrota 
man = mb.generace.mandelbrot(-2, 1.0, -1.5, 1.5, 1024, 100)
plt.imshow(man.reshape((1024,1024)))
```

Vypočti a vykresli Juliovu množinu:

```python
#vygeneruj Juliovu množinu 
jul = mb.generace.julia_set(-2, 1.0, -1.5, 1.5, -0.8 + 0.156j, 1024, 100)
plt.imshow(jul.reshape((1024,1024)))
```
